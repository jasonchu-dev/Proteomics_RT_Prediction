```python
import pandas as pd
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torch.utils.data import Dataset

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.cuda.empty_cache()

df = pd.read_csv('unmod.txt', sep='\t')
df
```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
df
```

```python
min_len, max_len = float('inf'), 0
for i in df['sequence']:
    min_len = len(i) if len(i) < min_len else min_len
    max_len = len(i) if len(i) > max_len else max_len

min_len, max_len, min(df['RT']), max(df['RT'])
```

```python
def normalize_neg_one_2_one():
    upper_bound = 10000
    normalized = (df['RT'] / upper_bound) - 1
    return normalized

def normalize_zero_2_one():
    upper_bound = 20000
    normalized = df['RT'] / upper_bound
    return normalized

df['RT'] = normalize_zero_2_one()
df['RT']
```

```python
X = df['sequence']
y = df['RT']
X_train, X_tmp, y_train, y_tmp = train_test_split(X, y, test_size=0.1, random_state=42)
X_test, X_val, y_test, y_val = train_test_split(X_tmp, y_tmp, test_size=0.5, random_state=42)
X_train.reset_index(drop=True, inplace=True)
X_val.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
y_train.reset_index(drop=True, inplace=True)
y_val.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)
```

```python
amino_acids = set(''.join(X_train))
idx2acid = dict(enumerate(sorted(amino_acids), start=1))
acid2idx = dict(map(reversed, idx2acid.items()))
```

```python
for seq in range(len(X_train)):
    sample = []
    for i in X_train[seq]:
        sample.append(acid2idx[i])
    sample.extend([0 for i in (range(max_len - len(sample)))])
    X_train[seq] = sample
    
for seq in range(len(X_val)):
    sample = []
    for i in X_val[seq]:
        sample.append(acid2idx[i])
    sample.extend([0 for i in (range(max_len - len(sample)))])
    X_val[seq] = sample
    
for seq in range(len(X_test)):
    sample = []
    for i in X_test[seq]:
        sample.append(acid2idx[i])
    sample.extend([0 for i in (range(max_len - len(sample)))])
    X_test[seq] = sample
```

```python
class CustomDataset(Dataset):
    def __init__(self, features, labels):
        # self.features = [torch.tensor(feature, dtype=torch.float) for feature in features]
        self.features = torch.tensor(features, dtype=torch.long)
        self.labels = torch.tensor(labels.values, dtype=torch.float)
        
    def __len__(self):
        return len(self.features)
    
    def __getitem__(self, idx):
        feature = self.features[idx]
        label = self.labels[idx]
        return feature, label
    
batch_size = int(len(X_train) / 4)
# batch_size = len(X_train)

train_data = torch.utils.data.DataLoader(
    CustomDataset(X_train, y_train),
    batch_size=batch_size,
    shuffle=False,
    drop_last=False,
    num_workers=0,
    pin_memory=False
)

val_data = torch.utils.data.DataLoader(
    CustomDataset(X_val, y_val),
    batch_size=batch_size,
    shuffle=False,
    drop_last=False,
    num_workers=0,
    pin_memory=False
)

test_data = torch.utils.data.DataLoader(
    CustomDataset(X_test, y_test),
    batch_size=batch_size,
    shuffle=False,
    drop_last=False,
    num_workers=0,
    pin_memory=False
)
```

```python
def unnormalize_neg_one_2_one(prediction):
    upper_bound = 10000
    unnormalized = (prediction + 1) * upper_bound
    return unnormalized

def unnormalize_zero_2_one(prediction):
    upper_bound = 20000
    unnormalized = prediction * upper_bound
    return unnormalized

def format_number(num, length=20):
    num_str = str(num)
    if "." in num_str:
        integer_part, decimal_part = num_str.split(".")
        formatted_decimal_part = decimal_part.ljust(3, '0')
        formatted_integer_part = integer_part
        formatted_num = "{}.{}".format(formatted_integer_part, formatted_decimal_part)
    else:
        formatted_num = num_str
    formatted_num = formatted_num.ljust(length, '0')
    return formatted_num
```

```python
class Encoder(nn.Module):
    def __init__(self, input_size=21, hidden_size=128, num_layers=2, embedding_size=10, dropout=0.3):
        super().__init__()
        self.num_layers = num_layers
        self.hidden_size = hidden_size
        self.dropout = nn.Dropout(dropout)
        self.embeddings = nn.Embedding(input_size, embedding_size)
        self.lstm = nn.LSTM(input_size=embedding_size, hidden_size=hidden_size, num_layers=num_layers)
        
    def forward(self, x):
        embedding = self.embeddings(x)
        embedding = self.dropout(embedding)
        output, (hn, cn) = self.lstm(embedding)
        return output, hn, cn
    
class Decoder(nn.Module):
    def __init__(self):
        super().__init__()
        # self.conv1d_1 = nn.Conv1d(38, 21, kernel_size=6, stride=2) # 64 -> 30
        # self.batchnorm_1 = nn.BatchNorm1d(21)
        # self.sigmoid_1 = nn.Sigmoid()
        
        # self.conv1d_2 = nn.Conv1d(21, 10, kernel_size=6, stride=3) # 30 -> 9
        # self.batchnorm_2 = nn.BatchNorm1d(10)
        # self.sigmoid_2 = nn.Sigmoid()
        
        # self.conv1d_3 = nn.Conv1d(10, 1, kernel_size=9, stride=1) # 9 -> 1
        # self.batchnorm_3 = nn.BatchNorm1d(1)
        # self.sigmoid_3 = nn.Sigmoid()
        
        # ------------------------

        # self.conv1d_1 = nn.Conv1d(38, 21, kernel_size=6, stride=3) # 96 -> 31
        # self.batchnorm_1 = nn.BatchNorm1d(21)
        # self.sigmoid_1 = nn.Sigmoid()
        
        # self.conv1d_2 = nn.Conv1d(21, 10, kernel_size=4, stride=3) # 31 -> 10
        # self.batchnorm_2 = nn.BatchNorm1d(10)
        # self.sigmoid_2 = nn.Sigmoid()
        
        # self.conv1d_3 = nn.Conv1d(10, 1, kernel_size=10, stride=1) # 10 -> 1
        # self.batchnorm_3 = nn.BatchNorm1d(1)
        # self.sigmoid_3 = nn.Sigmoid()

        # ------------------------

        self.conv1d_1 = nn.Conv1d(38, 24, kernel_size=6, stride=2) # 128 -> 61
        self.batchnorm_1 = nn.BatchNorm1d(24)
        self.sigmoid_1 = nn.Sigmoid()
        
        self.conv1d_2 = nn.Conv1d(24, 12, kernel_size=6, stride=5) # 61 -> 12
        self.batchnorm_2 = nn.BatchNorm1d(12)
        self.sigmoid_2 = nn.Sigmoid()

        self.conv1d_3 = nn.Conv1d(12, 1, kernel_size=12, stride=1) # 12 -> 1
        self.batchnorm_3 = nn.BatchNorm1d(1)
        self.sigmoid_3 = nn.Sigmoid()

        # ------------------------

        # self.conv1d_1 = nn.Conv1d(38, 19, kernel_size=12, stride=2) # 256 -> 123
        # self.batchnorm_1 = nn.BatchNorm1d(19)
        # self.sigmoid_1 = nn.Sigmoid()
        
        # self.conv1d_2 = nn.Conv1d(19, 1, kernel_size=123, stride=1) # 123 -> 1
        # self.batchnorm_2 = nn.BatchNorm1d(1)
        # self.sigmoid_2 = nn.Sigmoid()
        
    def forward(self, x):
        output = self.conv1d_1(x)
        output = self.batchnorm_1(output)
        output = self.sigmoid_1(output)
        
        output = self.conv1d_2(output)
        output = self.batchnorm_2(output)
        output = self.sigmoid_2(output)
        
        output = self.conv1d_3(output)
        output = self.batchnorm_3(output)
        output = self.sigmoid_3(output)
        
        return output
    
class peptide2RT(nn.Module):
    def __init__(self, encoder, decoder):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
        
    def forward(self, x):
        output, hn, cn = encoder(x)
        output = decoder(output)
        return output

lr = 0.1
weight_decay = 0.01
epochs = 2000

encoder = Encoder().to(device)
decoder = Decoder().to(device)
model = peptide2RT(encoder, decoder).to(device)
# criterion = nn.MSELoss(reduction='sum')
# # criterion = nn.SmoothL1Loss(reduction='sum', beta=0.5)
criterion = nn.HuberLoss(reduction='sum', delta=0.5)
optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=100, verbose=True, min_lr=1e-3)

torch.cuda.empty_cache()

for epoch in range(1, epochs+1):
    model.train()
    training_losses = []
    scaled_train_losses = []
    for feature, label in train_data:
        optimizer.zero_grad()
        y_pred = model(feature.to(device))
        y_pred = y_pred.squeeze(1, 2)
        training_loss = criterion(y_pred, label.to(device))
        scaled_train_loss = criterion(y_pred * 20000, (label * 20000).to(device))
        # training_loss.backward()
        scaled_train_loss.backward()
        optimizer.step()
        training_losses.extend([training_loss.item()])
        scaled_train_losses.extend([scaled_train_loss.item()])

    val_losses = []
    scaled_val_losses = []
    with torch.no_grad():
        for feature, label in val_data:
            y_pred = model(feature.to(device))
            y_pred = y_pred.squeeze(1, 2)
            val_loss = criterion(y_pred, label.to(device))
            scaled_val_loss = criterion(y_pred * 20000, (label * 20000).to(device))
            val_losses.extend([val_loss.item()])
            scaled_val_losses.extend([scaled_val_loss.item()])

    scheduler.step(scaled_train_loss)

    training_avg_loss = sum(training_losses) / len(training_losses)
    scaled_train_avg_loss = sum(scaled_train_losses) / len(scaled_train_losses)
    val_avg_loss = sum(val_losses) / len(val_losses)
    scaled_val_avg_loss = sum(scaled_val_losses) / len(scaled_val_losses)

    epoch = str(epoch).rjust(5)
    training_avg_loss = format_number(training_avg_loss)
    scaled_train_avg_loss = format_number(scaled_train_avg_loss)
    val_avg_loss = format_number(val_avg_loss)
    scaled_val_avg_loss = format_number(scaled_val_avg_loss)


    print(f'Epoch: {epoch} | Train Loss: {training_avg_loss} | Scaled Train Loss: {scaled_train_avg_loss} | Val Loss: {val_avg_loss} | Scaled Val Loss: {scaled_val_avg_loss}')
```

```python
def save_model(file_path, model, optimizer, train_loss, scaled_loss, epoch):
    checkpoint = {
        'epoch': epoch,
        'model_state': model.state_dict(),
        'optimizer_state': optimizer.state_dict(),
        'train_loss': train_loss,
        'scaled_loss': scaled_loss
    }
    torch.save(checkpoint, file_path)

save_model('./model.tar', model, optimizer, training_avg_loss, scaled_train_avg_loss, epoch)
```

```python
model.eval()

predictions = []
ground_truth = []

with torch.no_grad():
    for feature, label in test_data:
        y_pred = model(feature.to(device))
        y_pred = y_pred.squeeze(1, 2)
        y_pred = unnormalize_zero_2_one(y_pred.to('cpu'))
        label = unnormalize_zero_2_one(label)
        predictions.extend(y_pred)
        ground_truth.extend(label)

plt.figure(figsize=(25, 5))
plt.plot(predictions[:500])
plt.plot(ground_truth[:500])
plt.show()
```
