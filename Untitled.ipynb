```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset

df = pd.read_csv('unmod.txt', sep='\t')
df
```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
df
```

```python
min_len, max_len = float('inf'), 0
for i in df['sequence']:
    min_len = len(i) if len(i) < min_len else min_len
    max_len = len(i) if len(i) > max_len else max_len
min_len, max_len
```

```python
X = df['sequence']
y = df['RT']
X_train, X_tmp, y_train, y_tmp = train_test_split(X, y, test_size=0.1, random_state=42)
X_test, X_val, y_test, y_val = train_test_split(X_tmp, y_tmp, test_size=0.5, random_state=42)
X_train.reset_index(drop=True, inplace=True)
X_val.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
y_train.reset_index(drop=True, inplace=True)
y_val.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)
```

```python
amino_acids = set(''.join(X_train))
idx2acid = dict(enumerate(sorted(amino_acids), start=1))
acid2idx = dict(map(reversed, idx2acid.items()))
```

```python
for seq in range(len(X_train)):
    sample = []
    for i in X_train[seq]:
        sample.append(acid2idx[i])
    sample.extend([0 for i in (range(max_len - len(sample)))])
    X_train[seq] = sample
    
for seq in range(len(X_val)):
    sample = []
    for i in X_val[seq]:
        sample.append(acid2idx[i])
    sample.extend([0 for i in (range(max_len - len(sample)))])
    X_val[seq] = sample
    
for seq in range(len(X_test)):
    sample = []
    for i in X_test[seq]:
        sample.append(acid2idx[i])
    sample.extend([0 for i in (range(max_len - len(sample)))])
    X_test[seq] = sample
```

```python
class CustomDataset(Dataset):
    def __init__(self, features, labels):
        # self.features = [torch.tensor(feature, dtype=torch.float) for feature in features]
        self.features = torch.tensor(features, dtype=torch.long)
        self.labels = torch.tensor(labels.values, dtype=torch.float)
        
    def __len__(self):
        return len(self.features)
    
    def __getitem__(self, idx):
        feature = self.features[idx]
        label = self.labels[idx]
        return feature, label
    
batch_size = 1000

train_data = torch.utils.data.DataLoader(
    CustomDataset(X_train, y_train),
    batch_size=batch_size,
    shuffle=False,
    drop_last=False,
    num_workers=0,
    pin_memory=False
)

val_data = torch.utils.data.DataLoader(
    CustomDataset(X_val, y_val),
    batch_size=batch_size,
    shuffle=False,
    drop_last=False,
    num_workers=0,
    pin_memory=False
)

test_data = torch.utils.data.DataLoader(
    CustomDataset(X_test, y_test),
    batch_size=batch_size,
    shuffle=False,
    drop_last=False,
    num_workers=0,
    pin_memory=False
)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class Encoder(nn.Module):
    def __init__(self, input_size=21, hidden_size=64, num_layers=2, embedding_size=10, dropout=0.05, output_size=20):
        super().__init__()
        self.num_layers = num_layers
        self.hidden_size = hidden_size
        self.dropout = nn.Dropout(dropout)
        self.embeddings = nn.Embedding(input_size, embedding_size)
        self.lstm = nn.LSTM(input_size=embedding_size, hidden_size=hidden_size, num_layers=num_layers)
        
    def forward(self, x):
        embedding = self.embeddings(x)
        embedding = self.dropout(embedding)
        output, (hn, cn) = self.lstm(embedding)
        return output, hn, cn
    
class Decoder(nn.Module):
    def __init__(self):
        super().__init__()
        # self.model = nn.Sequential(
        #     nn.Conv1d(2, 1, kernel_size=6, stride=2), # 64 -> 30
        #     # nn.BatchNorm1d(2),
        #     nn.Sigmoid(),
        #     nn.Conv1d(1, 1, kernel_size=6, stride=3), # 30 -> 9
        #     # nn.BatchNorm1d(2),
        #     nn.Sigmoid(),
        #     nn.Conv1d(1, 1, kernel_size=9, stride=1), # 9 -> 1
        #     # nn.BatchNorm1d(2),
        #     nn.Sigmoid()
        # )
        
        self.conv1d_1 = nn.Conv1d(38, 21, kernel_size=6, stride=2) # 64 -> 30
        self.batchnorm_1 = nn.BatchNorm1d(21)
        self.sigmoid_1 = nn.Sigmoid()
        
        self.conv1d_2 = nn.Conv1d(21, 10, kernel_size=6, stride=3) # 30 -> 9
        self.batchnorm_2 = nn.BatchNorm1d(10)
        self.sigmoid_2 = nn.Sigmoid()
        
        self.conv1d_3 = nn.Conv1d(10, 1, kernel_size=9, stride=1) # 9 -> 1
        self.batchnorm_3 = nn.BatchNorm1d(1)
        self.sigmoid_3 = nn.Sigmoid()
        
        
    def forward(self, x):
        # print(x.shape)
        output = self.conv1d_1(x)
        output = self.batchnorm_1(output)
        output = self.sigmoid_1(output)
        
        output = self.conv1d_2(output)
        output = self.batchnorm_2(output)
        output = self.sigmoid_2(output)
        
        output = self.conv1d_3(output)
        output = self.batchnorm_3(output)
        output = self.sigmoid_3(output)
        
        return output
    
class peptide2RT(nn.Module):
    def __init__(self, encoder, decoder):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
        
    def forward(self, x):
        output, hn, cn = encoder(x)
        output = decoder(output)
        return output
    
lr = 1e-2
weight_decay = 0
dropout = 0    
epochs = 50

encoder = Encoder()
decoder = Decoder()
model = peptide2RT(encoder, decoder)
# criterion = nn.MSELoss(reduction='sum')
criterion = nn.SmoothL1Loss(reduction='mean', beta=1.0)
optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)

for epoch in range(1, epochs+1):
    model.train()
    losses = []
    # for i, (feature, label) in enumerate(train_data):
    for feature, label in train_data:
        optimizer.zero_grad()
        y_pred = model(feature)
        y_pred = y_pred.squeeze(1, 2)
        loss = criterion(y_pred, label.to(device))
        loss.backward()
        optimizer.step()
        losses.append(loss)
    avg_loss = sum(losses) / len(losses)
    print(f'Epoch: {epoch} --- Loss: {avg_loss}')
```

```python
model.eval()

predictions = []
ground_truth = []

with torch.no_grad():
    for feature, label in test_data:
        y_pred = model(feature)
        y_pred = y_pred.squeeze(1, 2)
        print(y_pred.shape, label.shape)
        predictions.extend(y_pred)
        ground_truth.extend(label)
```
